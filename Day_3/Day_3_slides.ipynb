# file I/O
- files are objects in Python
- form of getting input other than `input`, or `argv`
- great for storing small amounts of data without using a database
- What can you do?
  - open
    - read
    - write
    - append
  - close
  - create

# file I/O: open

`file_handle = open('filename.txt', 'r')`

- choose a name for the file handle so that we can work with it, call the open, pass a filename as a string, and the processing mode as a string, default is `'r'`
- processing modes:
  - `'r'` for read
  - `'w'` for write
  - `'a'` for append

# file I/O: close and create

- how do we close files?
  - `file_handle.close()`
  - try to close the files you open

- how do we create files?
  - open a file in write or append mode
  - `file_handle = open('filename.txt', 'w')`
  - create new file in the same directory as your script
    - can also specify the full path in the name

# file I/O: read

```bash
>>> f = open('read.txt', 'r')
```

What does `readlines()` do?

that signify the end of a line.

# file I/O: write

if write.txt exists, we will write over it:

``` python
f = open('write.txt', 'w’) # Create/open file
f.write(‘DSG’)
string = ‘ are my initials’
my_age = 29
f.write(string)
f.write(my_age)  # fails
f.write(str(my_age))
```

`write()` writes from wherever the file pointer is located

# file I/O: append
just like `write()`, but: starts with the file pointer at the end of the file (instead of the start).

``` python
f = open('write.txt', 'a')
f.write("THIS WAS APPENDED")
f.close()
```

``` python
f = open('write.txt', 'w')
f.write("ERR")
f.close()
```

# file I/O: with

``` python
with open("x.txt", "r") as f:
    data = f.read()
    do_something_with_data
```
